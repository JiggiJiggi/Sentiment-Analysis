### Preprocessing the text

```python
# Removing the unwanted symbols and punctuations that won't provide any info for our analysis
def text_preprocessing(text):
    text = re.sub(r'http\S+', '', text)  # Remove URLs
    text = re.sub(r'@\w+', '', text)  # Remove mentions
    text = re.sub(r'#\w+', '', text)  # Remove hashtags
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    text = text.lower()  # Convert to lowercase
    return text

# Apply the preprocessing function
df['text'] = df['text'].apply(text_preprocessing)
df.value_counts()
```

```python
# Unigrams and Bigrams Analysis
def get_top_ngrams(corpus, n=None, ngram_range=(1, 1)):
    vec = CountVectorizer(ngram_range=ngram_range).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
# Getting the most commonly used unigrams and bigrams
top_unigrams = get_top_ngrams(df['text'], n=20, ngram_range=(1, 1))
top_bigrams = get_top_ngrams(df['text'], n=20, ngram_range=(2, 2))
```

## Data Preparation

```python
# Shuffle the data
data = df.sample(frac=1, random_state=0).reset_index(drop=True)

# Filter the data to get 10000 entries for each class (0 and 4)
data_class_0 = data[data['sentiment'] == 0].sample(n=100000, random_state=0)
data_class_1 = data[data['sentiment'] == 4].sample(n=100000, random_state=0)

# Combine the samples and shuffle again
sampled_data = pd.concat([data_class_0, data_class_1]).sample(frac=1, random_state=0).reset_index(drop=True)
```

```python
# Preprocess the text data
X = sampled_data['text']
y = sampled_data['sentiment']

# Ensure y is a one-dimensional array
y = y.values.ravel()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Display shapes of the resulting datasets
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Feature Extraction

```python
# Convert text data into numerical data using TfidfVectorizer
tfidf_vector = TfidfVectorizer(max_features=50000, ngram_range=(1, 1))
X_train_tfidf = tfidf_vector.fit_transform(X_train)
X_test_tfidf = tfidf_vector.transform(X_test)


# Display shapes of the resulting datasets
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Model Training

#### Lets train few machine learning models to evaluate their performance

```python
from sklearn.naive_bayes import MultinomialNB # Importing Naive Bayes classifier for multinomially distributed data
from sklearn.linear_model import LogisticRegression # Importing Logistic Regression model 
from sklearn.ensemble import RandomForestClassifier # Importing Random Forest classifier for ensemble-based classification
from sklearn.metrics import classification_report, accuracy_score # for generating classification report and calculating accuracy score
from sklearn.svm import SVC #  Importing Support Vector Classifier for classification tasks
from sklearn.ensemble import GradientBoostingClassifier # Importing Gradient Boosting
```

##### Naive Bayes

```python
# Naive Bayes
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
nb_pred = nb_model.predict(X_test_tfidf)

# Print evaluation results for Naive Bayes
nb_accuracy = accuracy_score(y_test, nb_pred)
print(f'Naive Bayes Accuracy: {nb_accuracy}')
print(classification_report(y_test, nb_pred))
```

##### Logistic regression

```python
# Logistic Regression
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(X_train_tfidf, y_train)
lr_pred = lr_model.predict(X_test_tfidf)

# Print evaluation results for Logistic Regression
lr_accuracy = accuracy_score(y_test, lr_pred)
print(f'Logistic Regression Accuracy: {lr_accuracy}')
print(classification_report(y_test, lr_pred))
```

```python
# Random Forest
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_train_tfidf, y_train)
rf_pred = rf_model.predict(X_test_tfidf)

# Print evaluation results for Random Forest
rf_accuracy = accuracy_score(y_test, rf_pred)
print(f'Random Forest Accuracy: {rf_accuracy}')
print(classification_report(y_test, rf_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Gradient Boosting
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train_tfidf, y_train)
gb_pred = gb_model.predict(X_test_tfidf)

# Print evaluation results for Gradient Boosting
gb_accuracy = accuracy_score(y_test, gb_pred)
print(f'Gradient Boosting Accuracy: {gb_accuracy}')
print(classification_report(y_test, gb_pred))
```

## Model Evaluation

```python
# Evaluate models
models = {
    'Naive Bayes': nb_pred,
    'Logistic Regression': lr_pred,
    'Gradient Boosting Classifier': gb_pred,
    'Random Forest': rf_pred
}

for model_name, predictions in models.items():
    accuracy = accuracy_score(y_test, predictions)
    print(f'{model_name} Accuracy: {accuracy}')
    print(classification_report(y_test, predictions))

# Select the best model
accuracies = {model_name: accuracy_score(y_test, predictions) for model_name, predictions in models.items()}
best_model = max(accuracies, key=accuracies.get)
print(f'The best model is: {best_model}')
```

```python
text_user=[]
import uuid
class ChatBot:
    def __init__(self):
        # Stores conversation history, initialized empty
        self.conversations = {}

    def get_response(self, message):
        # Provides responses based on identified keywords
        keywords_responses = {
            "stress": "It's important to take breaks. What kind of work do you do?",
            "deadlines": "Breaking tasks into smaller goals can help. Do you enjoy your work?",
            "creative": "Working on projects you love can be rewarding. Have you tried gamifying your tasks?",
            "joke": "Why don't scientists trust atoms? Because they make up everything!",
            "help": "I'm here to help! What do you need assistance with?"
        }
        for keyword, response in keywords_responses.items():
            if keyword in message.lower():
                return response
        return "Can you share more about it?."

    def start_chat(self):
        print("Chatbot: Hi! How can I help you today? Type 'exit' to end the chat.")
        user_id = str(uuid.uuid4())  # Generates unique ID for the session
        self.conversations[user_id] = []

        while True:
            user_message = input("You: ")
            if user_message.lower() == 'exit':
                print("Chatbot: Goodbye! We can start sentiment analysis later.")
                break

            # Log user message
            self.conversations[user_id].append(("User", user_message))

            # Get chatbot response
            text_user.append(user_message)
            response = self.get_response(user_message)
            print(f"Chatbot: {response}")

            # Log chatbot response
            self.conversations[user_id].append(("Chatbot", response))

        # After chat ends, print the conversation history
        self.print_chat_history(user_id)

        return user_id

    def print_chat_history(self, user_id):
        print("\nChat History:")
        for speaker, message in self.conversations[user_id]:
            print(f"{speaker}: {message}")

# Example usage:
bot = ChatBot()
user_id = bot.start_chat()
```
